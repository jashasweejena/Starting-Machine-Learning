## value_counts() function returns the frequency of each point in the column, sorted in descending order from top to bottom

```python
#Top5 zip codes calling 911
dataFrame['zip'].value_counts().head(5)
```

```python
#Top5 TownShips calling 911
dataFrame['twp'].value_counts().head(5)
```

## nunique -> No. of unique values

```python
#Number of unique reasons to call 911
dataFrame['title'].nunique()
```

## .apply() function

## Helps to apply an operation on a whole column

```python
dataFrame['sp_rsn'] = dataFrame['title'].apply(lambda title: title.split(':')[1])
```

```python
dataFrame
```

```python
dataFrame['sp_rsn'].value_counts().head(20)
```

```python
dataFrame['rsn'] = dataFrame['title'].apply(lambda title: title.split(':')[0])
```

```python
dataFrame
```

```python
dataFrame['rsn'].value_counts().head(5)
```

## Countplot

```python
sns.countplot(x='rsn', data=dataFrame)
```
